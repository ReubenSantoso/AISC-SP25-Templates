```python
from fastapi import FastAPI
import uvicorn
from threading import Thread
from pyngrok import ngrok
```

```python
# Create the FastAPI app
app = FastAPI(title="AI-Enabled FastAPI with ngrok")

# Placeholder function where a beginner would call their AI model
def call_my_ai_model(input_data: dict) -> dict:
    """STUDENTS README: Replace this with your AI model inference code."""
    # Example stub: echo input
    return {"model_output": f"Echo: {input_data}"}
```

```python
#some sample calls you can do with a REST API. Read more about GET, POST, DELETE....
@app.post("/predict/")
def predict(payload: dict):
    """Endpoint for AI model inference"""
    # Call the beginner-student AI model hook
    result = call_my_ai_model(payload)
    return result
```

```python
# Function to run Uvicorn server
def run_server():
    uvicorn.run(app, host="0.0.0.0", port=8000)

# Start the server in a background thread
thread = Thread(target=run_server, daemon=True)
thread.start()

# Expose the server publicly via ngrok
public_url = ngrok.connect(8000)
print(f"Public URL: {public_url}")
```
